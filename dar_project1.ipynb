```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.style.use('ggplot')
```

```python
df=pd.read_csv("BTech_CS_DS_AI_3rdSem_Students_Cleaned_withNulls.csv")
display(df)
```

```python
print("Missing values in each column:")
print(df.isnull().sum())
print("Total missing values:",df.isnull().sum().sum())
```

```python
#to see in which row there is null values in attendance column
null_attendance_rows = df[df['Attendance'].isnull()]
display(null_attendance_rows)
```

```python
#to see the attendance column after filling null values
display(df['Attendance'])
df[df['Attendance'].isnull()]
```

```python
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
subjects = [
    "Discrete_Mathematics",
    "Software_Engineering",
    "Data_Analytics_and_Reporting",
    "Data_Structures_and_Algorithms",
    "Digital_Electronics",
    "Human_Values_and_Environmental_Studies"
]

for sub in subjects:
    df[sub].fillna(df[sub].mean(), inplace=True)
```

```python
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.chart import BarChart, Reference, LineChart, PieChart
```

```python
# --- Calculate Average Marks ---

subjects = [
    'Discrete_Mathematics',
    'Software_Engineering',
    'Data_Analytics_and_Reporting',
    'Data_Structures_and_Algorithms',
    'Digital_Electronics',
    'Human_Values_and_Environmental_Studies'
]
df["Average_Marks"] = df[subjects].mean(axis=1)
```

```python
# --- Create Attendance Groups ---

bins = [0, 60, 75, 90, 100]
labels = ["<60%", "60-75%", "75-90%", "90-100%"]
df["Attendance_Group"] = pd.cut(df["Attendance"], bins=bins, labels=labels, include_lowest=True)
```

```python
# --- Create Pivot Tables ---

pivot_gender = df.pivot_table(index="Gender", values="Average_Marks", aggfunc="mean").reset_index()
pivot_attendance = df.pivot_table(index="Attendance_Group", values="Average_Marks", aggfunc="mean").reset_index()
pivot_subject = df[subjects].mean().reset_index()
pivot_subject.columns = ["Subject", "Average_Marks"]
```

```python
# --- Create Excel Workbook ---

wb = Workbook()
ws1 = wb.active
ws1.title = "Gender_vs_AvgMarks"
```

```python
# --- Add Pivot: Gender vs Average Marks ---

for r in dataframe_to_rows(pivot_gender, index=False, header=True):
    ws1.append(r)
```

```python
# --- Bar Chart (Gender vs Avg Marks) ---

bar_chart = BarChart()
bar_chart.title = "Gender vs Average Marks"
bar_chart.x_axis.title = "Gender"
bar_chart.y_axis.title = "Average Marks"
data = Reference(ws1, min_col=2, min_row=1, max_row=ws1.max_row)
cats = Reference(ws1, min_col=1, min_row=2, max_row=ws1.max_row)
bar_chart.add_data(data, titles_from_data=True)
bar_chart.set_categories(cats)
ws1.add_chart(bar_chart, "E2")
```

```python
# --- Add Pivot: Attendance Group vs Average Marks ---

ws2 = wb.create_sheet("Attendance_vs_AvgMarks")
for r in dataframe_to_rows(pivot_attendance, index=False, header=True):
    ws2.append(r)
```

```python
# --- Pie Chart (Attendance Categories) ---

pie_chart = PieChart()
pie_chart.title = "Attendance Categories"
labels = Reference(ws2, min_col=1, min_row=2, max_row=ws2.max_row)
data = Reference(ws2, min_col=2, min_row=1, max_row=ws2.max_row)
pie_chart.add_data(data, titles_from_data=True)
pie_chart.set_categories(labels)
ws2.add_chart(pie_chart, "E2")
```

```python
# --- Add Pivot: Subject-wise Average ---

ws3 = wb.create_sheet("Subject_vs_AvgMarks")
for r in dataframe_to_rows(pivot_subject, index=False, header=True):
    ws3.append(r)
```

```python
# --- Line Chart (Subject-wise Average) ---

line_chart = LineChart()
line_chart.title = "Subject-wise Average Marks"
line_chart.x_axis.title = "Subject"
line_chart.y_axis.title = "Average Marks"
data = Reference(ws3, min_col=2, min_row=1, max_row=ws3.max_row)
cats = Reference(ws3, min_col=1, min_row=2, max_row=ws3.max_row)
line_chart.add_data(data, titles_from_data=True)
line_chart.set_categories(cats)
ws3.add_chart(line_chart, "E2")
```

```python
# --- Save the Excel File ---

output_file = "Student_Pivot_Report1.xlsx"
wb.save(output_file)

print(f"✅ Excel report '{output_file}' created successfully!")
```
